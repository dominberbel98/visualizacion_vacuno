El consumo de carne de vacuno ha bajado en muchos pases durante la ltima dcada por varios motivos interrelacionados:

Conciencia ambiental y cambio climtico
La produccin de carne de vacuno es una de las actividades ganaderas que genera ms emisiones de gases de efecto invernadero (metano, CO2). Por eso, cada vez ms gente reduce su consumo para minimizar su huella ambiental. Aunque como hemos visto en este estudio, eso no es del todo cierto, ya que existe una relacion directa pero leve.

Salud y bienestar
Campaas de salud pblica y estudios cientficos han vinculado el consumo excesivo de carne roja con riesgos para la salud, como enfermedades cardiovasculares y ciertos tipos de cncer. Esto ha fomentado dietas ms equilibradas o basadas en plantas.

Tendencias dietticas
El auge de dietas vegetarianas, veganas o flexitarianas ha hecho que muchas personas reduzcan el consumo de carne roja, sustituyndola por pollo, pescado, protenas vegetales o alternativas vegetales.

Precio y economa
En algunos mercados, la carne de vacuno suele ser ms cara que otras carnes (pollo, cerdo). Si los precios suben o hay crisis econmicas, el consumo de vacuno puede caer ms rpido.

Impacto de la produccin y oferta
Problemas en la cadena de suministro, regulacin ambiental o brotes de enfermedades en ganado (como la fiebre aftosa o la vaca loca en dcadas anteriores) pueden influir en la oferta y la demanda.

Cultura y moda gastronmica
La popularizacin de platos y cocinas que usan ms pollo o pescado, o la aparicin de superalimentos y productos saludables, influye en las preferencias del consumidor.
